```python
#可视化
def plot_results(model):
    spreads = [s["spreader"] for s in model.status_record]
    deads = [s["dead_end"] for s in model.status_record]
    reaches = [s["reached"] for s in model.status_record]

    plt.figure(figsize=(8, 4))
    plt.plot(spreads, label="Spreader (Sₜ)", marker="o", color="tomato")
    plt.plot(deads, label="Dead End (Dₜ)", marker="s", color="gray")
    plt.plot(reaches, label="Reached (Rₜ)", marker="^", color="purple")
    plt.title("News Diffusion Over Time")
    plt.xlabel("Day")
    plt.ylabel("Agent Count")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_results(model)
```
